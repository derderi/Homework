# Question 3

````python
def count1(dna, base):
    i = 0
    for c in dna:
        if c == base:
	    i += 1 
    return i

print 'count1(dna="asdfasdf", base="a"):', count1(dna="asdfasdf", base="a")
print 'count1(dna="asdfasdf", base="b"):', count1(dna="asdfasdf", base="b")
print 'count1(dna="asdfasdf", base="d"):', count1(dna="asdfasdf", base="d")
````

````python
def count2(dna, base):
    i = 0 
    for j in range(len(dna)):
        if dna[j] == base:
            i += 1 
    return i 

print 'count2(dna = "asdfasdf", base="a"):', count2(dna = "asdfasdf", base="a")
print 'count2(dna="asdfasdf", base="b"):', count2(dna="asdfasdf", base="b")
print 'count2(dna="asdfasdf", base="d"):', count2(dna="asdfasdf", base="d")
````

````python
def count3(dna, base):
        match = [c == base for c in dna]
        return sum(match)

print 'count3(dna = "asdfasdf", base="a"):', count3(dna = "asdfasdf", base="a")
print 'count3(dna="asdfasdf", base="b"):', count3(dna="asdfasdf", base="b")
print 'count3(dna="asdfasdf", base="d"):', count3(dna="asdfasdf", base="d")
````

````python
def count4(dna, base):
    return dna.count(base)

print 'count4(dna = "asdfasdf", base="a"):', count4(dna = "asdfasdf", base="a")
print 'count4(dna="asdfasdf", base="b"):', count4(dna="asdfasdf", base="b")
print 'count4(dna="asdfasdf", base="d"):', count4(dna="asdfasdf", base="d")
````

````python
def count5(dna, base):
    #print [i for i in range(len(dna)) if dna[i] == base]
    return len([i for i in range(len(dna)) if dna[i] == base])

print 'count5(dna = "asdfasdf", base="a"):', count5(dna = "asdfasdf", base="a")
print 'count5(dna="asdfasdf", base="b"):', count5(dna="asdfasdf", base="b")
print 'count5(dna="asdfasdf", base="d"):', count5(dna="asdfasdf", base="d")
````

````python
def count6(dna,base):
    #print [c == base for c in dna]
    #print (c == base for c in dna)
    return sum(c == base for c in dna)

print 'count6(dna = "asdfasdf", base="a"):', count6(dna = "asdfasdf", base="a")
print 'count6(dna="asdfasdf", base="b"):', count6(dna="asdfasdf", base="b")
print 'count6(dna="asdfasdf", base="d"):', count6(dna="asdfasdf", base="d")
````

# Where are the modules?
When a module is imported, Python first searches for a built-in module with that name.If a built-in module is not fount, Python then searches for a file obtained by adding the extension .py to the name of the module that it's imported:
1. in your current working directory,
2. the directory where Python has been installed,
3. in a path, i.e colon (':') separted list of file paths, stored in the environment variable PYTHONPATH.
